```python
from shapely.geometry import Point
import geopandas as gpd
```

```python
# Convert latitude and longitude columns to numeric
dist_centres[['latitude', 'longitude']] = dist_centres[['latitude', 'longitude']].apply(pd.to_numeric, errors='coerce')
users[['latitude', 'longitude']] = users[['latitude', 'longitude']].apply(pd.to_numeric, errors='coerce')
```

```python
# Create GeoDataFrames
centers_geometry = [Point(xy) for xy in zip(dist_centres['longitude'], dist_centres['latitude'])]
centers_gdf = gpd.GeoDataFrame(dist_centres, geometry=centers_geometry)
users_geometry = [Point(xy) for xy in zip(users['longitude'], users['latitude'])]
users_gdf = gpd.GeoDataFrame(users, geometry=users_geometry)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10, 10))
centers_gdf.plot(ax=ax, color='red', markersize=50, label='Distribution Centers')
users_gdf.plot(ax=ax, color='blue', markersize=5, label='Users')
plt.title('Geographic Distribution of Users and Distribution Centers')
plt.legend()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

#### Inference
1) The map shows the locations of distribution centers (marked in red) and users (marked in blue) on a world map.</br>
2) From the map, we can see where distribution centers are located relative to users. </br>
3) It appears that distribution centers are fewer in number and are spread out across different regions, while users are more densely distributed, especially in certain areas.</br>

```python
# Analysis
print("Analysis:")
print("Total distribution centers:", len(dist_centres))
print("Total users:", len(users))
print("Average latitude of distribution centers:", dist_centres['latitude'].mean())
print("Average longitude of distribution centers:", dist_centres['longitude'].mean())
print("Average latitude of users:", users['latitude'].mean())
print("Average longitude of users:", users['longitude'].mean())
```

```python
gdf_distribution_centers = gpd.GeoDataFrame(dist_centres,geometry=gpd.points_from_xy(dist_centres['longitude'],dist_centres['latitude']))
gdf_users = gpd.GeoDataFrame(users, geometry=gpd.points_from_xy(users['longitude'], users['latitude']))

# Create a basemap
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the basemap and the GeoDataFrames
fig, ax = plt.subplots(figsize=(10, 8))
world.plot(ax=ax, color='lightgrey')
gdf_distribution_centers.plot(ax=ax, color='red', label='Distribution Centers',
                              markersize=300)
gdf_users.plot(ax=ax, color='blue', label='Users', markersize=5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribution Centers and Users Locations')
plt.legend()
plt.show()
```

#### Inference
1) Users are heavily concentrated in North America, Europe, and Southeast Asia, as indicated by the blue dots.</br>
2) Distribution centers, represented by red dots, are strategically located near areas with high concentrations of users, indicating efficient placement to cater to user needs.</br>
3) There are sparse scattering of users in other areas where there are no nearby distribution centers, suggesting potential areas for expansion.</br>

**Geospatial Analysis**: By mapping the geographic distribution of users and distribution centers, i try gain insights into the spatial patterns of user engagement and logistical infrastructure. This analysis allows us to understand where users are located relative to distribution centers, enabling strategic decisions regarding inventory management, logistics planning, and market expansion.

# Order Fulfillment Analysis

```python
order_items.head()
```

```python
order = pd.read_csv('orders.csv')
order.head()
```

```python
order['status'].value_counts()
```

### Merge the data

```python
merged_df1 = pd.merge(order_items, order, on=['order_id', 'user_id'], how='left')
```

```python
merged_df1.head()
```

### Calculate fulfillment metrics

```python
merged_df1['order_created_at'] = pd.to_datetime(merged_df1['created_at_x'])
merged_df1['order_shipped_at'] = pd.to_datetime(merged_df1['shipped_at_x'])
merged_df1['order_delivered_at'] = pd.to_datetime(merged_df1['delivered_at_x'])
merged_df1['order_returned_at'] = pd.to_datetime(merged_df1['returned_at_x'])

merged_df1['shipping_time'] = merged_df1['order_shipped_at'] - merged_df1['order_created_at']
merged_df1['delivery_time'] = merged_df1['order_delivered_at'] - merged_df1['order_shipped_at']
merged_df1['return_time'] = merged_df1['order_returned_at'] - merged_df1['order_delivered_at']
```

```python
merged_df1.head()
```

### Visualize fulfillment timelines

```python
plt.figure(figsize=(12, 6))
sns.histplot(merged_df1['shipping_time'].dt.days, bins=20, kde=True)
plt.title('Distribution of Shipping Time')
plt.xlabel('Shipping Time (Days)')
plt.ylabel('Frequency')
plt.show()
```

#### Inference
1) The presence of negative values for shipping time suggest that some packages arrived earlier than expected.</br>
2) The most common delivery times are at -3, -1, 0, 1, and 2 days, as indicated by the peaks in the graph.</br>
3) There are troughs at approximately -2 and 3 days, suggesting these delivery times are less frequent.</br>

```python
plt.figure(figsize=(12, 6))
sns.histplot(merged_df1['delivery_time'].dt.days, bins=20, kde=True)
plt.title('Distribution of Delivery Time')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Frequency')
plt.show()
```

#### Inference
1) The delivery times are most frequently 0, 1, 2, 3, and 4 days, as indicated by the peaks in the blue bars.</br>
2) There seems to be a periodic pattern in the delivery times, as suggested by the oscillating line graph.</br>
3) The frequency of deliveries decreases at half-day intervals, reaching its lowest points between the peak days.</br>
4) Overall, the graph suggests that most deliveries are completed within 4 days, with certain days being more common for deliveries.

```python
plt.figure(figsize=(12, 6))
sns.histplot(merged_df1['return_time'].dt.days, bins=20, kde=True)
plt.title('Distribution of Return Time')
plt.xlabel('Return Time (Days)')
plt.ylabel('Frequency')
plt.show()
```

#### Inference
1) The return times are most frequently 0 and approximately 1.75 days, as indicated by the peaks in the blue bars.</br>
2) The graph shows a bimodal distribution of return times, suggesting there are two common periods when returns occur.</br>
3) The frequency of returns decreases between these two peaks.</br>
4) Overall, the graph suggests that most returns are completed within 2 days, with certain times being more common for returns.
